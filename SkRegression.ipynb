```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn import datasets, linear_model
```

```python
fishData = pd.read_csv(r'Fish.csv')
```

```python
#sort data

#y axis
breamWidthTrainData = np.transpose(fishData.iloc[0:30,1])
roachWidthTrainData = np.transpose(fishData.iloc[36:52,1])
whitefishWidthTrainData = np.transpose(fishData.iloc[56:59,1])
parkkiWidthTrainData = np.transpose(fishData.iloc[62:70,1])
perchWidthTrainData = np.transpose(fishData.iloc[73:120,1])
pikeWidthTrainData = np.transpose(fishData.iloc[129:142,1])
smeltWidthTrainData = np.transpose(fishData.iloc[146:157,1])

breamWidthTestData = np.transpose(fishData.iloc[31:35,1])
roachWidthTestData = np.transpose(fishData.iloc[53:55,1])
whitefishWidthTestData = np.transpose(fishData.iloc[60:61,1])
parkkiWidthTestData = np.transpose(fishData.iloc[71:72,1])
perchWidthTestData = np.transpose(fishData.iloc[121:128,1])
pikeWidthTestData = np.transpose(fishData.iloc[143:145,1])
smeltWidthTestData = np.transpose(fishData.iloc[158:159,1])

yfishWidthTrainData = np.hstack([breamWidthTrainData,roachWidthTrainData,whitefishWidthTrainData,parkkiWidthTrainData,perchWidthTrainData,pikeWidthTrainData,smeltWidthTrainData])
yfishWidthTestData = np.hstack([breamWidthTestData,roachWidthTestData,whitefishWidthTestData,parkkiWidthTestData,perchWidthTestData,pikeWidthTestData,smeltWidthTestData])


#x axis
breamTrainData = np.transpose(fishData.iloc[0:30,2:7]) #29
breamTestData = np.transpose(fishData.iloc[31:35,2:7]) #5
#breamTrainTarget = np.transpose(fishData.iloc[0:30,2:7]) #?
#breamTestTarget = np.transpose(fishData.iloc[0:30,2:7])

roachTrainData = np.transpose(fishData.iloc[36:52,2:7]) #16
roachTestData = np.transpose(fishData.iloc[53:55,2:7]) #3

whitefishTrainData = np.transpose(fishData.iloc[56:59,2:7]) #4
whitefishTestData = np.transpose(fishData.iloc[60:61,2:7]) #2

parkkiTrainData = np.transpose(fishData.iloc[62:70,2:7]) #8
parkkiTestData = np.transpose(fishData.iloc[71:72,2:7]) #2

perchTrainData = np.transpose(fishData.iloc[73:120,2:7]) #47
perchTestData = np.transpose(fishData.iloc[121:128,2:7]) #8

pikeTrainData = np.transpose(fishData.iloc[129:142,2:7]) #13
pikeTestData = np.transpose(fishData.iloc[143:145,2:7]) #3

smeltTrainData = np.transpose(fishData.iloc[146:157,2:7]) #11
smeltTestData = np.transpose(fishData.iloc[158:159,2:7]) #2

xfishTrainData = np.hstack([breamTrainData,roachTrainData,whitefishTrainData,parkkiTrainData,perchTrainData,pikeTrainData,smeltTrainData])
xfishTestData = np.hstack([breamTestData,roachTestData,whitefishTestData,parkkiTestData,perchTestData,pikeTestData,smeltTestData])

print(xfishTestData)
print(np.shape(xfishTestData))

#print(yfishWidthTestData)
print(np.shape(yfishWidthTestData))
```

```python
lengthOne = np.transpose(xfishTrainData[0:1,:])
lengthOneTest = np.transpose(xfishTestData[0:1,:])
#print(lengthOne)
#print(np.shape(lengthOne))
```

```python
#Length1

# Create linear regression object
regr = linear_model.LinearRegression()

# Train the model using the training sets
regr.fit(lengthOne, yfishWidthTrainData)

# The coefficients
print('Coefficients: \n', regr.coef_)
# The mean squared error
print("Mean squared error: %.2f"
      % np.mean((regr.predict(lengthOneTest) - yfishWidthTestData) ** 2))
# Explained variance score: 1 is perfect prediction
print('Variance score: %.2f' % regr.score(lengthOneTest, yfishWidthTestData))

# Plot outputs
plt.scatter(lengthOneTest, yfishWidthTestData,  color='black')
plt.plot(lengthOneTest, regr.predict(lengthOneTest), color='blue',
         linewidth=3)

plt.xticks(())
plt.yticks(())

plt.show()
```
